```python
def map_score(score):
    if score>=7:
        return 'High score'
    elif score<=3:
        return 'Low score'
    else:
        return 'Medium score'
```

```python
#map - pass a function that accepts input and returns output
df['Score'].map(map_score)
```

```python
df['Score'].map(map_score).value_counts()
```

```python
df['score_category'] = df['Score'].map(map_score)
df.head()
```
